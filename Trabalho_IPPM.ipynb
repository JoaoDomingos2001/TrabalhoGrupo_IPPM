#Teorema de Bolzano-Cauchy
O teorema de Bolzano-Cauchy diz que:

Dado uma função real de variável real $f$ contínua num intervalo $[a,b]$ em que $(a < b)$, para qualquer valor real de $k$ do intervalo de extremo $f(a)$ e $f(b)$, existe $c∈[a,b]$, tal que: $f(c)=k$

##Criação da função Bolzano-Cauchy

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a função
def f(x):
    return x**3 - 6*x + 4

# Parâmetros
a, b = -3, 3  # Intervalo [a, b]
k = 10        # Valor de k para a reta horizontal
x = np.linspace(a - 1, b + 1, 500)  # Intervalo de x para plotar a função
y = f(x)

# Verificar se o teorema de Bolzano-Cauchy se aplica
fa, fb = f(a), f(b)

if min(fa, fb) <= k <= max(fa, fb):
    print(f"O valor de k={k} está no intervalo [{min(fa, fb)}, {max(fa, fb)}].")
    print("O Teorema de Bolzano-Cauchy garante que existe c ∈ [a, b] tal que f(c) = k.")
else:
    print(f"O valor de k={k} não está no intervalo [{min(fa, fb)}, {max(fa, fb)}].")
    print("O Teorema de Bolzano-Cauchy não se aplica neste caso.")

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r"$f(x) = x^3 - 6x + 4$", color="blue")  # Função f(x)
plt.axhline(y=k, color="orange", linestyle="--", label=f"Reta y = {k}")  # Reta y=k
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Eixo x
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Eixo y

# Destaque nos pontos f(a) e f(b)
plt.scatter([a, b], [f(a), f(b)], color="red", s=100, label=f"Pontos: f(a)={fa}, f(b)={fb}")

# Região onde f(x) está acima e abaixo de k
plt.fill_between(x, y, k, where=(y > k), color="green", alpha=0.2, label="f(x) > k")
plt.fill_between(x, y, k, where=(y < k), color="red", alpha=0.2, label="f(x) < k")

# Configurações do gráfico
plt.title("Teorema de Bolzano-Cauchy para $f(x) = x^3 - 6x + 4$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

##Conclusão do Teorema de Bolzano-Cauchy
Para se utilizar este teorema é preciso:
*   função $f$ ser contínua no intervalo $[a,b]$;
*   e, o valor de $k$ está compreendido entre $f(a)$ e $f(b)$.

##Corolário do Teorema de Bolzano-Cauchy
Se $f$ é uma função contínua no intervalo $[a,b]$ e $f(a)$ e $f(b)$ têm sinais contrários, então a função $f$ tem pelo menos um zero pertencente a $]a,b[$.

Ou, dito de outro modo:

Se $f$ é contínua em $[a,b]$ e $f(a)*f(b)<0$ então, existe $c∈]a,b[:f(c)=0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a função
def f(x):
    return x**3 - 6*x + 4

# Parâmetros
a, b = -3, 3  # Intervalo [a, b]
x = np.linspace(a - 1, b + 1, 500)  # Intervalo de x para plotar a função
y = f(x)

# Função para verificar o corolário e criar o gráfico
def plot_corolario(a, b):
    fa, fb = f(a), f(b)
    print(f"f(a) = {fa}, f(b) = {fb}")

    # Verificar condição do corolário
    if fa * fb < 0:
        print(f"As condições do corolário são satisfeitas: f(a) * f(b) < 0.")
        print("Portanto, existe pelo menos um c ∈ ]a, b[ tal que f(c) = 0.")
    else:
        print("As condições do corolário não são satisfeitas: f(a) * f(b) >= 0.")
        print("Não é garantido que exista um zero no intervalo ]a, b[.")

    # Gráfico
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=r"$f(x) = x^3 - 6x + 4$", color="blue")  # Função f(x)
    plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # Eixo x
    plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # Eixo y
    plt.scatter([a, b], [fa, fb], color="red", s=100, label=f"Pontos: f(a)={fa}, f(b)={fb}")

    # Destacar a região onde o zero está garantido
    if fa * fb < 0:
        c_approx = (a + b) / 2  # Ponto médio como uma aproximação do zero
        plt.scatter([c_approx], [f(c_approx)], color="green", s=100, label="Aproximação do zero")

    # Configurações do gráfico
    plt.title("Corolário do Teorema de Bolzano-Cauchy", fontsize=14)
    plt.xlabel("x", fontsize=12)
    plt.ylabel("f(x)", fontsize=12)
    plt.legend()
    plt.grid()
    plt.show()

# Chamar a função para o corolário
plot_corolario(a, b)
```

#### Método da Bissecção
Baseia-se do Corolário do Teorema de Bolzano-Cauchy, para aplicar é preciso fazer estes seguintes passos:



1.   Verificação do Corolário, ou seja, se $f$ é contínua em $[a,b]$ e $f(a)*f(b)<0$ então, existe $c∈]a,b[:f(c)=0$.

2.   Calcula-se o ponto médio $c = (a+b)/2$

3.   Testa-se o ponto médio, $f(c)$:
      *   Se f(c) = 0, c é raiz da solução
      *   Se f(c) != 0, determinamos em que subintervalo $[a,c]$ ou $[c,b]$ está a raiz ao aplicar outra vez o corolário: $f(a)*f(c)<0$

4.   Repetir o processo:
      *   Atualizar o intervalo para qual contém a raiz $[a,c]$ ou $[c,b]$
      *   Continue dividindo o intervalo pela metade até que o comprimento do intervalo seja menor que um limite predefinido (𝜖) ou até atingir um número máximo de iterações.



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a função
def f(x):
    return x**3 - 6*x + 4

# Método da Bissecção
def bisseccao(a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("O Corolário de Bolzano não é aplicável: f(a) * f(b) >= 0.")

    print(f"Iterando o Método da Bissecção no intervalo [{a}, {b}]")
    iteracoes = []
    raiz = None

    for i in range(max_iter):
        c = (a + b) / 2
        iteracoes.append((a, b, c, f(c)))
        print(f"Iteração {i+1}: a={a:.6f}, b={b:.6f}, c={c:.6f}, f(c)={f(c):.6f}")

        # Verifica se encontrou a raiz
        if abs(f(c)) < tol:  # f(c) próximo de zero
            raiz = c
            print("Convergência alcançada!")
            break

        # Determina o novo intervalo
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

        # Critério de parada baseado no comprimento do intervalo
        if abs(b - a) < tol:
            raiz = c
            print("O comprimento do intervalo é menor que a tolerância.")
            break

    if raiz is not None:
        print(f"A raiz aproximada encontrada é: c ≈ {raiz:.6f}")
    else:
        print("Não foi possível determinar uma raiz com os parâmetros fornecidos.")

    return iteracoes, raiz

# Função para visualizar o Método da Bissecção
def plot_bisseccao(iteracoes, raiz=None):
    x = np.linspace(-4, 4, 500)
    y = f(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=r"$f(x) = x^3 - 6x + 4$", color="blue")
    plt.axhline(0, color="black", linewidth=0.8, linestyle="--")

    # Destaque cada iteração no gráfico
    for i, (a, b, c, fc) in enumerate(iteracoes):
        plt.scatter(c, fc, color="red", label=f"Iteração {i+1}" if i == 0 else "")
        plt.plot([a, b], [f(a), f(b)], "g--", alpha=0.5)

    # Destaque da raiz
    if raiz is not None:
        plt.scatter(raiz, 0, color="purple", s=100, label=f"Raiz aproximada: c ≈ {raiz:.6f}")

    plt.title("Método da Bissecção", fontsize=14)
    plt.xlabel("x", fontsize=12)
    plt.ylabel("f(x)", fontsize=12)
    plt.legend()
    plt.grid()
    plt.show()

# Parâmetros
a, b = -3, 3  # Intervalo inicial
tol = 1e-6    # Tolerância
max_iter = 30  # Número máximo de iterações

# Executar o Método da Bissecção
iteracoes, raiz = bisseccao(a, b, tol, max_iter)

# Plotar o gráfico do Método da Bissecção
plot_bisseccao(iteracoes, raiz)
```
